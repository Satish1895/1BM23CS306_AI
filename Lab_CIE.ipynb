```python
# Step 1: Initialize Knowledge Base

facts = set([
    ("HolidaySpot", "Darjeeling"),
    ("HighlyVisited", "Munar"),
    ("HighlyVisited", "Darjeeling"),
    ("HighlyVisited", "It"),
])

# It is not Munar
not_equal = [("It", "Munar")]

# Step 2: Define inference rules
# Rule: If a place is a HolidaySpot -> IceCreamAvailable
def rule_holiday_spot(facts):
    new_facts = set()
    for (pred, x) in facts:
        if pred == "HolidaySpot":
            new_facts.add(("IceCreamAvailable", x))
    return new_facts

# Rule: If It is a HighlyVisited place and there are only two HighlyVisited (Munar, Darjeeling)
# and It != Munar, then It = Darjeeling
def rule_identify_it(facts, not_equal):
    new_facts = set()
    hv_places = [x for (pred, x) in facts if pred == "HighlyVisited"]
    for (a, b) in not_equal:
        if a == "It" and b in hv_places:
            for p in hv_places:
                if p != b:
                    new_facts.add(("IdentifiedAs", "Darjeeling"))
                    new_facts.add(("HolidaySpot", "It"))
    return new_facts

#  Step 3: Forward chaining

def forward_chain(facts, not_equal):
    inferred = set()
    changed = True
    while changed:
        changed = False
        new1 = rule_identify_it(facts, not_equal)
        new2 = rule_holiday_spot(facts)

        new_facts = new1 | new2
        new_facts = new_facts - facts

        if new_facts:
            changed = True
            facts |= new_facts
            inferred |= new_facts

    return facts, inferred


#  Step 4: Run reasoning
final_facts, inferred = forward_chain(facts, not_equal)

# Step 5: Check if IceCreamAvailable(It) is proven
icecream_available = ("IceCreamAvailable", "It") in final_facts

print("Final Facts:")
for f in final_facts:
    print(" ", f)

if icecream_available:
    print("\n Proven: Ice-creams are available at 'It'.")
else:
    print("\n Could not prove the query.")

```
